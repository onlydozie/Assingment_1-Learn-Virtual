```python
#Covid19 analysis using dataframs and append method for new values

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fig = plt.figure(figsize= (100,50))
data = {'Date': ['29/04/2020','30/04/2020','01/05/2020', '02/05/2020', '03/05/2020', '04/05/2020', '05/05/2020', 
                 '06/05/2020', '07/05/2020', '08/05/2020', '09/05/2020', '10/05/2020'],
        'Number_of_Cases':[1728,1932,2170,2388,2558,2802,2950,3145,3526,3912,4151,4399], 
        'No_of_recoveries': [307,317,351,351,400,417,481,534,601,679,745,778], 
        'No_of_deaths':[51,58,68,85,87,93,98,103,107,117,128,143]}

Data = pd.DataFrame(data)

appendInput = input ('Any new data today? \n(y) yes \n(n) no')
if (appendInput.lower() == 'y'):        
    data2 = {'Date': '29/04/2020','Number_of_Cases':1728,'No_of_recoveries': 307,'No_of_deaths':51}
    value1 = input('Enter the date:')
    value2 = int(input('Enter the number of cases:'))
    value3 = int(input('Enter the number of recoveries:'))
    value4 = int(input('Enter the number of deaths:'))
    data2['Date'] = value1
    data2['Number_of_Cases'] = value2
    data2['No_of_recoveries'] = value3
    data2['No_of_deaths'] = value4

    Data = Data.append(data2, ignore_index=True)
else :
    Data = pd.DataFrame(data)

x = Data.Date
y = Data.Number_of_Cases
z = Data.No_of_recoveries
w = Data.No_of_deaths

plt.plot(x,y, marker='v', c = 'm', linewidth=7, markersize =20)
plt.plot(x,z, marker='v', c = 'g', linewidth=7, markersize =20)
plt.plot(x,w, marker='v', c = 'r', linewidth=7, markersize =20)
plt.title('Nigeria\'s COVID -19 Curve', fontsize = 100)
plt.xlabel('Timeline', fontsize=100)
plt.ylabel('Confirmed Cases No of recoveries No of death', fontsize=100)
plt.legend(['Confirmed Cases','Recoveries', 'Deaths'],loc='best', borderaxespad=0, fontsize = 100)
plt.xticks(fontsize = 70, rotation =90)
plt.grid(True)
plt.yticks(fontsize = 80)
plt.show()

Data
```
