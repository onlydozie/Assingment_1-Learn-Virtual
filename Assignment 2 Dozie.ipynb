```python
def prime_numbers():
    prime_check = int(input('Please enter a number:'))
    prime = True
    for primer in range (2, 10):
        if prime_check % primer == 0 and prime_check != primer:
            prime = False
    if prime is True:
            print ('It is a prime number')
    else:
            print ('It is not a prime number')
    
```
