```python
import matplotlib
import matplotlib.pyplot as plt #Loading Matplotlib plotting tools
import pandas as pd
import numpy as np
%matplotlib inline

x = [1, 3,5,9, 14, 21, 30, 40, 55]
y = [1,2,3,4,5,6,7,8,9]
plt.hist(x, bins=20, color='m')
plt.xlabel('X-Axis', fontsize=16)
plt.ylabel('Y-axis', fontsize=16)
plt.title('A simple Histogram', fontsize=16)
```

```python
#Replicating the plots using object oriented syle
x = [1, 3,5,9, 14, 21, 30, 40, 55]
y = [1,2,3,4,5,6,7,8,9]
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y, c='m')
#ax.set(title='A plot of X Vs Y', xlabel='X-axis',ylabel='Y-axis')
ax.legend(['X vs Y'],loc=('best'), borderaxespad=0, fontsize = 20)
ax.set_title('A plot of X vs Y', fontsize = 20)
ax.set_xlabel('X-axis', fontsize=16)
ax.set_ylabel('Y-axis', fontsize=16)
```

```python
x = [1, 3,5,9, 14, 21, 30, 40, 55]
y = [1,2,3,4,5,6,7,8,9]
z = [9,8,7,6,5,4,3,2,1]
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y,marker = 'v', c='m')
ax.plot(x,z, marker = '<', c='c')
#ax.set(title='A plot of X Vs Y', xlabel='X-axis',ylabel='Y-axis')
ax.legend(['X vs y', 'X vs z'],loc=('best'), borderaxespad=0, fontsize = 13)
ax.set_title('A plot of X vs Y and X vs z', fontsize = 20)
ax.set_xlabel('X-axis', fontsize=16)
ax.set_ylabel('Y-axis and Z-axis', fontsize=16)
```

```python
#Scatter plot using object oriented syle

from numpy.random import rand
x, y, scale = rand(3, 2000)
scale = 50 * scale ** 5
fig = plt.figure()
ax = plt.axes()
ax.scatter(x,y, s=scale, c='g')
#ax.set(title='A plot of X Vs Y', xlabel='X-axis',ylabel='Y-axis')
ax.legend(['X vs Y'],loc=(1,0.5), borderaxespad=0, fontsize = 20)
ax.set_title('A simple Scatter Plot', fontsize = 20)
ax.set_xlabel('X-axis', fontsize=16)
ax.set_ylabel('Y-axis', fontsize=16)
```

```python
# BarCharts from Categorcial Columns using object oriented syle

data = {'States': ['Lagos', 'Rivers', 'Delta', 'Oyo', 'Imo', 'Kano', 'Edo'],        
        'GDP':[33679,21073,16749,16121,14212,12393,11888]}
Data = pd.DataFrame(data)
Data.head()

plt.style.use('seaborn')
x = Data.States
y = Data.GDP
fig = plt.figure()
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid(True)
ax = plt.axes()
ax.bar(x,y, color='black')
#ax.set(title='A plot of X Vs Y', xlabel='X-axis',ylabel='Y-axis')
ax.legend(['Multiply GDP by 1,000,000'],loc=('best'), borderaxespad=0, fontsize = 13)
ax.set_title('States in Nigeria with the highest in 2010', fontsize = 20)
ax.set_xlabel('State', fontsize=16)
ax.set_ylabel('GDP', fontsize=16)
```

```python
# Pie chart. Anticlockwise order 

data = {'Tutors': ['Aminah', 'Adeola', 'Fatimah', 'Stephen', 'Femi', 'Isreal', 'Olayinka'],
        'Participants': ['joseph', 'Benjamin', 'Emmanuel', 'Kafee', 'Onyinye', 'Dominique', 'Andiwa'],
        'tutors_gender': ['Female', 'Female', 'Female', 'Male', 'Male', 'Male', 'Male'],
        'participant_gender': ['Male', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female'], 
        'Age':[25,29,24,34,32,27,30]}
Data = pd.DataFrame(data)
Data.head()

labels = Data.Tutors
sizes = Data.Age
explode = (0, 0.1, 0, 0,0,0,0) # only "explode" the 2nd slice



fig1, ax1 = plt.subplots() #matplotlib subplot() function can be called to plot two or more plots in one figure.e.g 2X1 2X2
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', 
        shadow=True, startangle=90) 
plt.show()
```
