```python
# Daily Activity

weekend = ['Saturday', 'Sunday']
churchday = ['Monday', 'Thursday', 'Wednesday']
Akaraday = ['Tuesday', 'Friday']
Spending_limit = 2000
day = input('What day is it today?')
bank_balance = int(input('Please enter your bank balance'))
if (day in churchday):
    print ('There is church service today')
elif (day in Akaraday) and (bank_balance >= Spending_limit):
    print ('You can eat but Akara today')
elif (day in Akaraday) and (bank_balance <= Spending_limit):
    print ('Sorry, you cannot eat Akara today')
else :
    print ('You can play video game all day')
```
