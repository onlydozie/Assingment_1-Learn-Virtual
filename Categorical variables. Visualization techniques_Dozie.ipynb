```python
#Categorical variables. Visualization techniques


import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
Ice_Cream_Shop = ['New York','LA','San Francisco']
Units_Sold = [12327,17129,19923]
df = pd.DataFrame(list(zip(Ice_Cream_Shop,Units_Sold)),columns=['Ice Cream Shop','Units Sold'])
total = sum (Units_Sold)
Ice_Cream_Shop1 = ['Total']
Units_Sold1 = [total]
df2 = pd.DataFrame(list(zip(Ice_Cream_Shop1,Units_Sold1)),columns=['Ice Cream Shop','Units Sold'])
df3 = df.append(df2, ignore_index=True)
x = Ice_Cream_Shop
y = Units_Sold
plt.bar(x,y, color='black')
plt.legend(['Units Sold'],loc=(1.02,0), borderaxespad=0, fontsize = 13)
plt.title('Barchat of Ice cream sold', fontsize = 20)
plt.xlabel('States', fontsize=10, )
plt.ylabel('Frequency', fontsize=10)
plt.xticks(fontsize = 10, rotation =0)
plt.grid(True)
plt.yticks(fontsize = 10)
df3
```

```python
plt.style.use('classic')
#plt.figure(figsize=(3,3))
plt.barh(x,y, color='black', align ='center', alpha =1)
plt.xlabel('States', color ='black', size = 20)
plt.ylabel('Frequency', color ='black', size = 20)
plt.title('Barchat of Ice cream sold', color = 'black', size = 20)
plt.legend(['Units Sold'],loc=(1.02,0), borderaxespad=0, fontsize = 13)
plt.xticks(fontsize=15,rotation='horizontal')
plt.yticks(fontsize=15,rotation='horizontal')
for index,value in enumerate(y):
    plt.text(value,index,str(value),fontsize=20)
plt.gca().invert_yaxis()
plt.show()
df3
```

```python
labels = Ice_Cream_Shop
sizes = Units_Sold
explode = (0, 0, 0.1)



fig1, ax1 = plt.subplots()
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', 
        shadow=True, startangle=90) 
plt.show()
```
